```python
df=pd.read_csv('./data/auto-mpg.csv',header=None)

df.columns=['mpg','cylinders','displacement','horsepower','weight','acceleration','model year','origin','name'] # 열 이름 지정

df.head() # 데이터 살펴보기
```

```python
df.describe() # 숫자형 데이터의 요약정보를 통해 이상치를 탐색
```

```python
df.isnull().sum() # 결측값이 있는지 확인
```

```python
df.duplicated().sum() # 각 행에 대해서 중복되는 행이 있는지 확인
```

```python
sns.pairplot(df); # 각 변수들 간의 관계와 분포(상관계수) 확인 -> 알부 변수들에 대해서 선형관계가 확인됨
```

```python
corr=df.corr(numeric_only=True) # 상관계수 분석
corr
```

```python
mask=np.triu(np.ones_like(corr,dtype=bool))
# np.triu : 행렬의 위쪽 삼각형 부분만 남기고 나머지는 False로 변환 (대칭성에 의해 필요없는 데이터)
# np.ones_like : corr와 같은 모양의 True로 채운 행렬 생성

plt.figure(figsize=(10,8))
sns.heatmap(corr,
            mask=mask,
            cmap='coolwarm',
            annot=True, # 각 셀에 상관계수 값 표시
            fmt=".2f", # 상관계수 값을 소수점 둘째 자리까지 표시
            cbar=True, # 오른쪽에 색상 막대 표시
            square=True) # 셀을 정사각형 모양으로 표시

plt.show()
```

```python
df['horsepower'].unique()
```

```python
df['horsepower']=df['horsepower'].replace('?',np.nan)
df['horsepower']=df['horsepower'].astype('float')
df.describe()
```

```python
print(df['horsepower'].isnull().sum())
df_nan=df.dropna(subset=['horsepower'],axis=0)
print(df_nan['horsepower'].isnull().sum())
```

```python
print(df['horsepower'].isnull().sum())
df['horsepower']=df['horsepower'].fillna(df['horsepower'].mean())
print(df['horsepower'].isnull().sum())
```

```python
corr=df.corr(numeric_only=True)
corr
```

```python
ndf=df[['mpg','cylinders','horsepower','weight']]
ndf.head()
```

```python
ndf.plot(kind='scatter',x='weight',y='mpg',c='coral',s=10,figsize=(10,5))
plt.show()
```

```python
fig=plt.figure(figsize=(10,5))
ax1=fig.add_subplot(1,2,1)
ax2=fig.add_subplot(1,2,2)

sns.regplot(x='weight',y='mpg',data=ndf,ax=ax1)
sns.regplot(x='weight',y='mpg',data=ndf,ax=ax2,fit_reg=False)
plt.show()
```

```python
sns.jointplot(x='weight',y='mpg',data=ndf)
sns.jointplot(x='weight',y='mpg',kind='reg',data=ndf)
```

```python
X=ndf[['weight']]
y=ndf[['mpg']]

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X, # 독립 변수
                                               y, # 종속 변수
                                               test_size=0.3, # 검증 30%
                                               random_state=10) # 랜덤 추출 값
print('train data 개수 : ',len(X_train))
print('test data 개수 : ',len(X_test))
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()

lr.fit(X_train,y_train)

r_square=lr.score(X_test,y_test)
print('R^2 결정계수 : ',r_square)
```

```python
print('기울기 a : ',lr.coef_)
print('y절편 b : ',lr.intercept_)
```

```python
y_hat=lr.predict(X_test)

test_preds=pd.DataFrame(y_test)
test_preds.columns=['y_test']
test_preds['y_hat']=y_hat
test_preds['squared_error']=(test_preds['y_hat']-test_preds['y_test'])**2
test_preds
```

```python
mse=test_preds['squared_error'].mean()
print('mse : ',mse)
```

```python
fig,axes=plt.subplots(1,2,figsize=(10,5))
sns.regplot(x='y_test',y='y_hat',data=test_preds,ax=axes[0]);
sns.kdeplot(x='squared_error',data=test_preds,ax=axes[1]);
```
