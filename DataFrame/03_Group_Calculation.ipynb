```python
import pandas as pd
import seaborn as sns
```

```python
titanic=sns.load_dataset('titanic')

df=titanic.loc[:,['age','sex','class','fare','survived']]

df.head()
```

```python
grouped=df.groupby(['class'],observed=True) # observed는 범주형 열에서 나타나지 않는 범주를 제외할지 여부를 결정
print(grouped)
```

```python
for key,group in grouped:
    print('* key : ',key)
    print('* number : ',len(group))
    print(group.head())
    print('\n')
```

```python
average=grouped.mean(numeric_only=True)
average
```

```python
group3=grouped.get_group(('Third',))
group3.head()
```

```python
grouped_two=df.groupby(['class','sex'],observed=True)

for key,group in grouped_two:
    print('* key : ',key)
    print('* number : ',len(group))
    print(group.head())
    print('\n')
```

```python
average_two=grouped_two.mean(numeric_only=True)
average_two
```

```python
group3f=grouped_two.get_group(('Third','female'))
group3f.head()
```

```python
titanic=sns.load_dataset('titanic')
df=titanic.loc[:,['age','sex','class','fare','survived']]

grouped=df.groupby(['class'],observed=True)

std_all=grouped.std(numeric_only=True) # 표준편차
std_all
```

```python
titanic=sns.load_dataset('titanic')
df=titanic.loc[:,['age','sex','class','fare','survived']]

grouped=df.groupby(['class'],observed=True)

agg_grouped=grouped[['age','survived']].apply(lambda x:x.describe())
agg_grouped
```

```python
def z_score(x):
    return (x-x.mean())/x.std()

age_zscore=grouped[['age','survived']].apply(z_score)
age_zscore
```

```python
age_filter=grouped[['age','survived']].apply(lambda x:x['age'].mean()<30)
print(age_filter,'\n')

df.loc[df['class'].isin(age_filter[age_filter==True].index),['age','survived']]
```
