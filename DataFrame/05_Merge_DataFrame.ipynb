```python
import pandas as pd
import requests
```

```python
df1=pd.DataFrame({'a':['a0','a1','a2','a3'],
                  'b':['b0','b1','b2','b3'],
                  'c':['c0','c1','c2','c3']},
                 index=[0,1,2,3])
df1
```

```python
df2=pd.DataFrame({'a':['a2','a3','a4','a5'],
                  'b':['b2','b3','b4','b5'],
                  'c':['c2','c3','c4','c5'],
                  'd':['d2','d3','d4','d5']},
                 index=[2,3,4,5])
df2
```

```python
result1=pd.concat([df1,df2])
result1
```

```python
result2=pd.concat([df1,df2],ignore_index=True)
result2
```

```python
result2_in=pd.concat([df1,df2],axis=0,join='inner') # inner 옵션은 교집합에 해당하는 열만 결과에 포함
result2_in
```

```python
result3=pd.concat([df1,df2],axis=1)
result3
```

```python
result3_in=pd.concat([df1,df2],axis=1,join='inner') # inner 옵션은 교집합에 해당하는 행만 결과에 포함
result3_in
```

```python
df1=pd.read_excel('./data/stock_price.xlsx')
df1
```

```python
df2=pd.read_excel('./data/stock_valuation.xlsx')
df2
```

```python
merge_inner=pd.merge(df1,df2,how='inner',on='id')
merge_inner
```

```python
merge_outer=pd.merge(df1,df2,how='outer',on='id')
merge_outer
```

```python
merge_left=pd.merge(df1,df2,how='left',left_on=['id','stock_name'],right_on=['id','name'])
merge_left
```

```python
merge_right=pd.merge(df1,df2,how='right',left_on=['id','stock_name'],right_on=['id','name'])
merge_right
```

```python
merge_cross=pd.merge(df1,df2,how='cross')
merge_cross
```

```python
sdf1=pd.DataFrame({'employee':['Alice','Sam','Eva'],
                   'department':['HR','Tech','HR']})
sdf2=pd.DataFrame({'employee':['Eva','Alice','Sam'],
                   'start_year':[2018,2019,2020]})

result_one_to_one=pd.merge(sdf1,sdf2,on='employee')
result_one_to_one
```

```python
sdf3=pd.DataFrame({'department':['HR','Tech'],
                   'manager':['Tina','Alex']})

result_many_to_one=pd.merge(sdf1,sdf3,on='department')
result_many_to_one
```

```python
sdf4=pd.DataFrame({'department':['HR','HR','Tech','Tech','Finance'],
                   'task':['recruiting','payroll','development','support','budgeting']})

result_many_to_many=pd.merge(sdf1,sdf4,on='department')
result_many_to_many
```

```python
my_key="api_key"

def get_financials(stockCode,indicatorName,apiKey,consolidated=True,ttm=True):
    if consolidated:
        if ttm:
            url=f"https://api.dapada.io/company/getConsolidatedFinancialDataByTTM?indicatorName={indicatorName}&stockCode={stockCode}"
        else:
            url=f"https://api.dapada.io/company/getConsolidatedFinancialDataByCUR?indicatorName={indicatorName}&stockCode={stockCode}"
    else:
        if ttm:
            url=f"https://api.dapada.io/company/getSeparatedFinancialDataByTTM?indicatorName={indicatorName}&stockCode={stockCode}"
        else:
            url=f"https://api.dapada.io/company/getSeparatedFinancialDataByCUR?indicatorName={indicatorName}&stockCode={stockCode}"

    headers={"Authorization":f"{apiKey}"}
    response=requests.get(url,headers=headers)
    result=response.json()
    return pd.DataFrame(result)

df_revenue=get_financials(stockCode='005930',
                          indicatorName='매출액',
                          apiKey=my_key,
                          consolidated=True, # 연결 재무제표 조회 여부
                          ttm=False) # Trailing Twelve Months : 최근 12개월간의 누적 데이터를 조회할지를 설정

df_revenue.columns=['매출액','분기']
df_revenue=df_revenue.set_index('분기',drop=True) # '분기' 열을 인덱스로 설정하고 해당 열을 삭제
df_revenue=df_revenue[df_revenue.index.str.startswith('2023')]

df_revenue
```

```python
df_margin=get_financials(stockCode='005930',
                         indicatorName='영업이익',
                         apiKey=my_key,
                         consolidated=True,
                         ttm=False)

df_margin.columns=['영업이익','분기']
df_margin=df_margin.set_index('분기',drop=True)
df_margin=df_margin[df_margin.index.str.endswith('Q3')]

df_margin
```

```python
df_revenue_margin=df_revenue.join(df_margin)
df_revenue_margin
```

```python
df_revenue_margin2=df_revenue.join(df_margin,how='inner')
df_revenue_margin2
```
