```python
import pandas as pd
import seaborn as sns
```

```python
titanic=sns.load_dataset('titanic')
df=titanic.loc[0:9,['age','fare']]
df
```

```python
df.query('age<20')
```

```python
df.query('age>=20 and fare<15')
```

```python
df.query('age>=20 or fare<15')
```
