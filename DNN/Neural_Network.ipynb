```python
import tensorflow as tf
from tensorflow import keras
```

```python
print(dir(tf.keras.activations))
```

```python
import numpy as np
import matplotlib.pylab as plt

# -10부터 10까지 0.1 간격의 숫자를 배열에 담음 (-10~9.9)
x=np.arange(-10,10,0.1)
```

```python
# sigmoid 함수에 배열 x를 통과시킴
# σ(x)=1/(1+e^(-x)) : 출력값을 0과 1 사이로 압축
tf.keras.activations.sigmoid(x)
```

```python
plt.plot(x,tf.keras.activations.sigmoid(x),linestyle='--',label='linear')
plt.legend()
```

```python
# ReLU 함수 : 입력값이 0 이하일 경우 0, 0보다 클 경우 입력값을 그대로 반환
tf.keras.activations.relu(x)
```

```python
plt.plot(x,tf.keras.activations.relu(x),linestyle='--',label='linear')
plt.legend()
```

```python
model=tf.keras.models.Sequential([ # 레이어를 순차적으로 쌓아 구성된 신경망 제작
    tf.keras.layers.Flatten(input_shape=(28,28)), # 입력 데이터의 차원을 변환 (28x28 픽셀의 2D 배열 -> 28*28 형태의 1D 배열)
    tf.keras.layers.Dense(128,activation='relu'), # 128개의 뉴런으로 구성된 연결층. 활성화 함수로 ReLU를 사용해 비선형성 추가
    tf.keras.layers.Dense(10,activation='softmax') # 마지막 레이어, 10개의 뉴런. Softmax 활성화 함수는 각 클래스의 확률을 출력
])
```

```python
model.summary()
```
