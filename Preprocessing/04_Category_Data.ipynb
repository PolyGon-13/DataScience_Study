```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
df=pd.read_csv('./data/auto-mpg.csv',header=None)

df.columns=['mpg','cylinder','displacement','horsepower','weight','acceleration','model year','origin','name']

df['horsepower']=df['horsepower'].replace('?',np.nan)
df=df.dropna(subset=['horsepower'],axis=0)
df['horsepower']=df['horsepower'].astype('float')

count,bin_dividers=np.histogram(df['horsepower'],bins=3)

print(count) # 구간에 속하는 값의 개수
print(bin_dividers) # 경계값 리스트
```

```python
bin_names=['저출력','보통출력','고출력']

df['hp_bin']=pd.cut(x=df['horsepower'], # 데이터 배열
                    bins=bin_dividers, # 경계값 리스트
                    labels=bin_names, # bin 이름
                    include_lowest=True) # 첫 경계도 포함

df[['horsepower','hp_bin']].head(15)
```

```python
count,bin_dividers=np.histogram(df['horsepower'],bins=3)

bin_names=['저출력','보통출력','고출력']

df['hp_bin']=pd.cut(x=df['horsepower'],
                    bins=bin_dividers,
                    labels=bin_names,
                    include_lowest=True)

# 컴퓨터만 인식할 수 있도록 0과 1로만 구성되는 원핫벡터로 변환하는 과정을 원핫인코딩이라고 함
horsepower_dummies=pd.get_dummies(df['hp_bin']) # 원핫인코딩
print(horsepower_dummies.head(15))
```

```python
horsepower_dummies_float=pd.get_dummies(df['hp_bin'],dtype=float)
horsepower_dummies_float.head()
```

```python
horsepower_dummies_drop=pd.get_dummies(df['hp_bin'],dtype=float,drop_first=True)
horsepower_dummies_drop.head()
```

```python
label_encoder=preprocessing.LabelEncoder() # 범주형 데이터 -> 숫자형 데이터 변환 함수
onehot_encoder=preprocessing.OneHotEncoder() # 2차원 범주형 데이터 -> 원핫벡터 변환 함수

onehot_labeled=label_encoder.fit_transform(df['hp_bin'].head(15))
print(onehot_labeled)
print(type(onehot_labeled))

onehot_reshaped=onehot_labeled.reshape(len(onehot_labeled),1) # nx1 형태로 변환 (2차원 데이터로 변환)
print(onehot_reshaped)
print(type(onehot_reshaped))

onehot_fitted=onehot_encoder.fit_transform(onehot_reshaped)
print(onehot_fitted)
print(type(onehot_fitted))
```
