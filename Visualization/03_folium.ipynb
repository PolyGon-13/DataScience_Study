```python
df=pd.read_excel('./data/서울지역_대학교_위치.xlsx',index_col=0)

seoul_map=folium.Map(location=[37.55,126.98],tiles="Cartodb Positron",zoom_start=12)

for name,lat,lng in zip(df.index,df['위도'],df['경도']): # zip : 여러 시퀀스(리스트, 배열)를 묶어서 동시에 순회할 수 있도록 함
    folium.Marker([lat,lng],popup=name).add_to(seoul_map)
    # popup : 마커를 클릭할 때 표시되는 팝업 정보 지정
    # add_to : 생성한 마커를 seoul_map이라는 지도 객체에 추가

seoul_map
```

```python
seoul_map2=folium.Map(location=[37.55,126.98],tiles="Cartodb Positron",zoom_start=12) # seoul_map을 그대로 사용하면 위의 기본마커와 겹쳐서 출력됨

for name,lat,lng in zip(df.index,df['위도'],df['경도']):
    folium.CircleMarker([lat,lng],
                        radius=10,
                        color='brown',
                        fill=True,
                        fill_color='coral',
                        fill_opacity=0.7,
                        popup=name
                        ).add_to(seoul_map2)

seoul_map2
```

```python
file_path='./data/경기도인구데이터.xlsx'
df=pd.read_excel(file_path,index_col='구분')
df.columns=df.columns.map(str) # 열 이름들을 문자열형으로 변환

geo_path='./data/경기도행정구역경계.json'
try:
    geo_data=json.load(open(geo_path,encoding='utf-8'))
except:
    geo_data=json.load(open(geo_path,encoding='utf-8-sig'))
'''
attr=('&copy; <a href="https://www.stadiamaps.com/" target="_blank">Stadia Maps</a> &copy; <a href="https://www.stamen.com/" target="_blank">Stamen Design</a> &copy; <a href="https://openmaptiles.org/" target="_blank">OpenMapTiles</a> &copy; <a href="https://www.openstreetmap.org/copyright">OpenStreetMap</a> contributors')
tiles='https://tiles.stadiamaps.com/tiles/stamen_terrain/{z}/{x}/{y}{r}.png'
'''

# 위 코드는 에러가 발생하여 다른 타일 사용
attr=('&copy; <a href="https://www.openstreetmap.org/copyright">OpenStreetMap</a> contributors')
g_map=folium.Map(location=[37.5502,126.982],tiles='OpenStreetMap',attr=attr,zoom_start=9)

year='2007'
folium.Choropleth(geo_data=geo_data, # 지도의 경계 정보를 담고 있는 GeoJSON 형식의 데이터
                 data=df[year], # 색상을 입힐 데이터
                 columns=[df.index,df[year]], # 데이터프레임의 두 개의 열을 지정 (첫 번째 열은 식별자, 두 번째 열은 색상으로 나타낼 값)
                 fill_color='YlOrRd',fill_opacity=0.7,line_opacity=0.3, # 색상을 설정하는 옵션, 색상의 불투명도 설정, 색상의 단계 설정
                 threshold_scale=[10000,100000,300000,500000,700000], # 색상의 단계(스케일)을 설정
                 key_on='feature.properties.name'  # 데이터의 어떤 속성을 기준으로 데이터를 매칭할지 설정
                 ).add_to(g_map)

g_map
```
