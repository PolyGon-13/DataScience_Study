```python
unique_values=df['origin'].value_counts() # 고유값의 개수
unique_values
```

```python
unique_values_ratio=df['origin'].value_counts(normalize=True) # 고유값의 상대적인 비율
unique_values_ratio
```

```python
unique_values_percentage=(df['origin'].value_counts(normalize=True)*100).round(1) # 비율을 백분율로 변환
unique_values_percentage
```
