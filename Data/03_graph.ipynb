```python
import pandas as pd
```

```python
df=pd.read_excel('./data/남북한발전전력량.xlsx')

df_ns=df.iloc[[0,5],3:]

df_ns.index=['South','North']

df_ns=df_ns.astype(int)

df_ns.head()
```

```python
df_ns.info()
```

```python
df_ns.plot()
```

```python
tdf_ns=df_ns.T
print(tdf_ns.head())
print('\n')
tdf_ns.plot(); # 세미콜론 붙이면 그래프 출력과정에서 잡다한 출력 생략하고 그래프만 출력함
```

```python
tdf_ns=df_ns.T
print(tdf_ns.head())
print('\n')
tdf_ns.plot(kind='bar');
```

```python
tdf_ns.plot(kind='barh');
```

```python
tdf_ns.plot(kind='bar',stacked=True);
```

```python
tdf_ns=df_ns.T
tdf_ns.plot(kind='hist');
```

```python
tdf_ns.plot(kind='hist',alpha=0.5,figsize=(6,2),grid=True,color='skyblue',edgecolor='black');
```

```python
tdf_ns.hist(figsize=(6,2));
```

```python
tdf_ns.hist(bins=20,grid=False,figsize=(6,2));
```
